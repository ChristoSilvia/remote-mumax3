```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('perturb2.1.txt').T
nice_data_intermediate = data.T[data[0,:] > 2.5e-9].T
nice_data = nice_data_intermediate.T[nice_data_intermediate[4,:] == 1e10].T

def filter_freq(alldata,freq):
    freq_index = 4
    return alldata.T[alldata[freq_index] == freq].T

def last_half(alldata):
    return alldata.T[len(alldata)/2:].T

def get_amplitude(alldata,freq):
    freq_index = 4
    osc_index = 2
    data_plus_initial = filter_freq(alldata,freq)
    data = last_half(data_plus_initial)
    
    return np.abs(np.max(data[osc_index]) - np.mean(data[osc_index]))

plt.plot(data[0],-data[2],label="m_y")
plt.plot(data[0],1e-11*data[4], label="1e-11*freq")
plt.legend()
plt.xlabel("Time (sec)")

print "Frequency | Amplitude"
for freq in [1e10,1.1e10,1.2e10,1.3e10,1.4e10,1.5e10,1.6e10]:
    print "{0}  {1}".format(freq, get_amplitude(data,freq))
```

```python
freqs = [1e10,1.1e10,1.2e10,1.3e10,1.4e10,1.5e10,1.6e10]
plt.plot(freqs, [get_amplitude(data,freq) for freq in freqs])
plt.title("Preliminary resonance peak")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Oscillation amplitude")
```
