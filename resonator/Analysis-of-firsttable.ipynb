### The 16-cell-deep experiment

This is the top of the data file I got from mumax3.  The code used to produce it is at the bottom of the page.  This is a 128x128x16 sample.

```python
datafile16 = open('resonance-16layer.txt')
for i in range(0,2):
    print datafile16.readline()
```

So the y-data is the third column, `data[2]`.  We can extract and plot it, along with the other data, to see which one looks the best.

```python
%pylab inline
import numpy as np
import pylab

data16 = np.loadtxt('resonance-16layer.txt').T


pylab.plot(data16[0],data16[1], label="m_x")
pylab.plot(data16[0],data16[2], label="m_y")
pylab.plot(data16[0],data16[3], label="m_z")
pylab.plot(data16[0],np.sin(-2*np.pi*1e9*data16[0]), label="B_y")
pylab.legend()
```

The $m_y$ is oscillating at a period of about 1ns.  That looks like what we want to focus in on.  Let's remove the other components of the magnetization.

```python
%pylab inline
import numpy as np
import pylab

data16 = np.loadtxt('resonance-16layer.txt').T

pylab.plot(data16[0],data16[2], label="m_y")
pylab.plot(data16[0],-np.sin(2*np.pi*1e9*data16[0]), label="B_y")
pylab.legend()
```

The magnetization appears to lag behind the external field, but suddenly change.  Notice the many vertical drops, where the magnetiztion suddenly moves a lot closer to the field.

## The 2 cell deep experiment

The above simulation took _4 hours_ to run.  I tried reducing the cell depth to 2, and it now takes _20 minutes_.  Let's see if it procuces similar output.  Note that I added in a field for the B_y in the official datatable, instead of with post-processing.

```python
datafile2 = open('resonance-2layer.txt')
for i in range(0,2):
    print datafile2.readline()
```

```python
%pylab inline
import numpy as np
import pylab

data2 = np.loadtxt('resonance-2layer.txt').T

pylab.plot(data2[0],data2[2], label="m_y")
pylab.plot(data2[0],0.66*data2[5], label="B_y")
pylab.legend()
```

At first glance these outputs look extremely similar.  But, I want to do a more detailed analysis of the differences.  What is a good way of figuring things out about periodic data?  FOURIER TRANSFORMS!!!

```python
%pylab inline 

diff = np.fft.ifft(np.fft.fft(data2[2][:(len(data16[2]))]) - np.fft.fft(data16[2]))
pylab.plot(data16[0],np.real(diff))
pylab.plot(data16[0],data16[2])
pylab.plot(data2[0][:(len(data16[2]))],data2[2][:(len(data16[2]))])
```

These are acceptable similar.
